#### Empty cells

```python
import pandas as pd

df = pd.read_csv('data.csv')

print(df)
```

```python
# cleaning
new_df = df.dropna()

print(new_df)
```

```python
# to change the original DataFrame, use the inplace = True argument

df.dropna(inplace=True)
print(df)
```

```python
# fillna() method allows us to replace empty cells with a value
data = pd.read_csv('data.csv')
n_df = data.fillna(130, inplace=True)
print(data)
```

```python
# Replace Only For Specified Columns

dframe = pd.read_csv('data.csv')

dframe['Calories'].fillna(130, inplace=True)

print(dframe)
```

#### Replace Using Mean, Median, or Mode

```python
# Mean =  the average value (the sum of all values divided by number of values).
dframe = pd.read_csv('data.csv')

mean_value = dframe['Calories'].mean()

dframe['Calories'].fillna(mean_value, inplace=True)

print(dframe)
```

```python
# Median = the value in the middle, after you have sorted all values ascending.

dframe = pd.read_csv('data.csv')

median_value = dframe['Calories'].median()

dframe['Calories'].fillna(median_value, inplace=True)

print(dframe)
```

```python
# Mode = the value that appears most frequently.

dframe = pd.read_csv('data.csv')

mode_value = dframe['Calories'].mode()[0]

dframe['Calories'].fillna(mode_value, inplace=True)

print(dframe)
```
